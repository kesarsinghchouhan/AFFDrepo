# datacleaning

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df
```

```python
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
# remove the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns name
df.columns=df.columns.str.strip()
df.columns
```

## change the requied columns as integer data type

day,month,year,Temperature,RH,WS

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

# changing the other columns datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
### lets save the cleaned data set
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv',index=False)
```

# Exploratory Data Analysis

```python
# drop day,month,year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy["Classes"].unique()
```

```python
df_copy['Classes']=df["Classes"].str.strip()
```

```python
df_copy['Classes'].unique()
```

```python
df_copy["Classes"]=df_copy["Classes"].map({'fire':1,'not fire':0})
```

```python
df_copy.head()
```

```python
df_copy["Classes"].value_counts()
```

# visualizations

```python
# plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
plt.style.available
```

```python
## percentage for pie chart
percentages_classes=df_copy["Classes"].value_counts(normalize=True)*100  # normalize will change it into percentage
percentages_classes
```

```python
# plotting pie chart
classlabels=['fire','not fire']
plt.figure(figsize=(20,7))
plt.pie(percentages_classes,labels=classlabels,autopct='%1.1f%%',shadow=True)
plt.title(" pie chart of classes")
```
